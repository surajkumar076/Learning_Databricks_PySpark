```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()
data = [('James','Smith','M',3000),
  ('Anna','Rose','F',4100),
  ('Robert','Williams','M',6200), 
]

columns = ["firstname","lastname","gender","salary"]
df = spark.createDataFrame(data=data, schema = columns)
df.show()
```

```
# This function calls for each partition
def reformat(partitionData):
    for row in partitionData:
        yield [row.firstname+","+row.lastname,row.gender,row.salary*10/100]

df2=df.rdd.mapPartitions(reformat).toDF(["name","gender","bonus"])
df2.show()
```

```
# Another method
def reformat(partitionData):
  updatedData = []
  for row in partitionData:
    name=row.firstname+","+row.lastname
    gender=row.gender
    bonus=row.salary*10/100
    updatedData.append([name,gender,bonus])
  return iter(updatedData)

df2=df.rdd.mapPartitions(reformat).toDF(["name","gender","bonus"])
df2.show()
```
