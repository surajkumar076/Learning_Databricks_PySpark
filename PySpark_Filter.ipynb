```
# Create DatFrame
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()

data = [("James","edy","Smith","36636","M",6000),
        ("Michael","Rose","wills","40288","M",7000),
        ("Robert","colm","Williams","42114","M",4000),
        ("Maria","Anne","Jones","39192","F",5000),
        ("Jen","Mary","Brown","56342","F",3000)]

columns = ["first_name","middle_name","last_name","dob","gender","salary"]
pysparkDF = spark.createDataFrame(data = data, schema = columns)

display(pysparkDF)
```

```
display(pysparkDF.filter(pysparkDF.salary <=5000))
```

```
display(pysparkDF)
display(pysparkDF.filter(pysparkDF.gender =="F"))
```

```
display(pysparkDF.filter(pysparkDF.last_name.like("%wills")))
```
