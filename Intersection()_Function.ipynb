```
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()

Fruit_Data = [("Watermelon","green","round"),
             ("Mango","yellow","oval"),
             ("Papaya","orange","oval"),
             ("Strawberry","red","round")]

columns= ["fruit_name","color","shape"]
df = spark.createDataFrame(data = Fruit_Data, schema = columns)
df.show()

Fruit_Data2 = [("Blackberry","black","round"),
              ("Apple","red","round"),
              ("Watermelon","green","round"),
              ("Strawberry","red","round")]
columns2= ["fruit_name","color","shape"]
df1 = spark.createDataFrame(data = Fruit_Data2, schema = columns2)
df1.show()
```

```
# Intersect of two dataframe in pyspark

df_inter=df.intersect(df1)
df_inter.show()
```

```
# intersect of more than two tables
 
from functools import reduce
from pyspark.sql import DataFrame
 
def intersect(*dfs):
    return reduce(DataFrame.intersect, dfs)
 
intersect(df, df1).show() #intersect(df, df1, df_nth_number........).show()
```
